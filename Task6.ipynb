### **Importing necessary libraries required**

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
```

### **Reading from the csv dataset as pandas dataframe (downloaded through above mentioned link)**

```python
df = pd.read_csv('Iris.csv')
df.head()  #Displaying first 5 values
```

### **Converting species datatype (i.e.,object) to integer and storing in variable "Species_n"**

```python
le_species = LabelEncoder()
df['Species_n'] = le_species.fit_transform(df['Species'])
df.head()    #Displaying first 5 values
```

### **Preparing the data**
**1. Storing the required data columns into variables "input" and "target".**

```python
input = df.drop(['Id','Species','Species_n'],axis=1)
target = df['Species_n']
```

```python
input.head()
```

**2. Splitting the data into two sets i.e.,training and testing**

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(input,target,test_size=0.2,random_state=0)
```

### **Training the model**

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
model.fit(X_train,Y_train)
print('Decision Tree Classifier Created')
```

### **Measuring the accuracy of model against the training data**

```python
model.score(X_train,Y_train)
```

### **Making Predictions**

```python
model.predict([[5.0,3.6,1.4,0.2]])
```

### **Visualizing Decision Tree**

```python
plt.figure(figsize=(30,20))
tree.plot_tree(model,filled=True,feature_names=input.columns,rounded=True,)
plt.show()
```
